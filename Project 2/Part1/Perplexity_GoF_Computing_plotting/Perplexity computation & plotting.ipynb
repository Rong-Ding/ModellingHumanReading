```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# read in the files
df_lstm = pd.read_csv('surprisal.lstm.test_set.csv')
df_srn = pd.read_csv('surprisal.srn.test_set.csv')
```

```python
# compute average surprisal (i.e., mean surprisal) by corpus size
mean_lstm = df_lstm.mean().tolist()[2:]
mean_srn = df_srn.mean().tolist()[2:]
```

```python
print(len(mean_lstm))
```

```python
# compute perplexity
x = [2]*9
perp_lstm = np.power(x, mean_lstm)
perp_srn =np.power(x, mean_srn)
```

```python
# plotting: lstm and srn represented in different colors, corpus size as x-axis
models = ["SRN", "LSTM"]
steps = ['1k','3k','10k','30k','100k','300k','1M','3M','8.7M']
```

```python
l1 = plt.plot(steps,perp_srn,'r',marker='o',label='SRN')
l2 = plt.plot(steps,perp_lstm,'b',marker='x',label='LSTM')

#plt.figure(figsize=(3, 3))
plt.xlabel('Training Corpus Size')
plt.ylabel('Perplexity')
plt.xticks(steps)
plt.legend()

plt.savefig("Perplexity", bbox_inches = "tight")
plt.show()
```
